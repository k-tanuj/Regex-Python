STRING METHODS


1. len(): provides length of any given string


```python
name ="tanuj"

print(len(name))
```

2. upper(): capitalize whole string

```python
upper = name.upper()
print(upper)
```

3. lower(): convert whole string to lower case 

```python
lower = name.lower()
print(lower)
```

4. isupper(): checks if the whole string is capitalized or not.

```python
name1 = "tanuj"
name2 = "Tanuj"
name3 = "TANUJ"

print(name1.isupper())
print(name2.isupper())
print(name3.isupper())
```

5. islower() : checks if all char are in lower case or not

```python
name1 = "tanuj"
name2 = "Tanuj"
name3 = "TANUJ"

print(name1.islower())
print(name2.islower())
print(name3.islower())
```

Question: Take string from user and print if it has more than 10 char or not.

```python
string = input("enter string")

if len(string) > 10:
    if string.isupper():
        print("more than 10 and upper")
    elif string.islower():
        print("more than 10 and lower")
    else:
        print("more than 10 and mixed")
else:
     if string.isupper():
        print("less than 10 and upper")
     elif string.islower():
        print("less than 10 and lower")
     else:
        print("less than 10 and mixed")
```

Question: 4 people are playing atlas.1st said india. 2nd said australia. 3rd said zimbabwe and 4th said england. check who said wrong.
      

```python
guy1 = input("guy1: ").lower()
guy2 = input("guy2: ").lower()
guy3 = input("guy3: ").lower()
guy4 = input("guy4: ").lower()

if guy1[-1] != guy2[0]:
    print("guy2 incorrect")
elif guy2[-1] != guy3[0]:
    print("guy3 incorrect")
elif guy3[-1] != guy4[0]:
    print("guy4 incorrect")
else:
    print("all correct")
```
