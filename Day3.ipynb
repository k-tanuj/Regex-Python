If the year is leap year


```python
year = int(input())
if year%4==0 and year%100!=0:
    print("year is leap")
elif year%400==0:
    print("year is leap")
else:
    print("not leap year")
```

Basic calculator

```python
num1 = int(input("enter 1st no.:"))
num2 = int(input("enter 2nd no.:"))
operator = input("enter operator:")

if operator == '+':
    print(num1+num2)
elif operator =='-':
    print(num1-num2)
elif operator == '*':
    print(num1*num2)    
elif operator =='/':
    print(num1/num2)
else:
    print("invalid operator")
```

Calculate BMI

```python
wt = float(input("enter weight in kg"))
ht = float(input("enter height in meters"))

BMI = wt/(ht**2)

if BMI < 18.5:
    print(f"your bmi is {BMI}: underweight")
elif BMI>=18.5 and BMI<=24.9:
    print(f"your bmi is {BMI}: normal")
elif BMI>=25 and BMI<=29.9:
    print(f"your bmi is {BMI}: OverWT")
else:
    print(f"your bmi is {BMI}: Obesity")
```
