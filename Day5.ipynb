String Slicing: Selecting multiple chars from a string

```python
name = "Tanuj Kumawat"
len(name)
print(name[0:5])
print(name[:5])
print(name[6:13])
print(name[6:])
print(name[:])
```

```python
# [ start index : end index : step]
print(name[::2])
print(name[0:13:2])
```

```python
#slicing with negative indexings
print(name[-1:-10:-1]) #tawamuK j
print(name[-1:-10:-2]) #twmKj

#all index provided should be either positive or negative

#below one's will not give output because step is positive and we are going from right to left
print(name[-1:-10:1])
print(name[1:10:-1])
print(name[-1:10:1])
```
